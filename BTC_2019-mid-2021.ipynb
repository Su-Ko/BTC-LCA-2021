# full LCA 2019

```python
mylist = [('IPCC 2013', 'climate change', 'GWP 100a'),
         ('ReCiPe Midpoint (H) (obsolete)', 'climate change', 'GWP100'),
         ('ReCiPe Midpoint (H) (obsolete)', 'freshwater ecotoxicity', 'FETPinf'),
         ('ReCiPe Midpoint (H) (obsolete)', 'freshwater eutrophication', 'FEP'),
         ('ReCiPe Midpoint (H) (obsolete)', 'human toxicity', 'HTPinf'),
         ('ReCiPe Midpoint (H) (obsolete)', 'marine ecotoxicity', 'METPinf'),
         ('ReCiPe Midpoint (H) (obsolete)', 'metal depletion', 'MDP'),
         ('ReCiPe Midpoint (H) (obsolete)', 'ozone depletion', 'ODPinf'),
         ('ReCiPe Midpoint (H) (obsolete)', 'photochemical oxidant formation', 'POFP'),
         ('ReCiPe Midpoint (H) (obsolete)', 'terrestrial acidification', 'TAP100'),
         ('ReCiPe Midpoint (H) (obsolete)', 'terrestrial ecotoxicity', 'TETPinf'),
         ('ReCiPe Midpoint (H) (obsolete)', 'fossil depletion', 'FDP')]
```

```python
def dolcacalc(myact_19, mydemand, mymethod):
    my_fu = {myact_19: mydemand} 
    lca = LCA(my_fu, mymethod)
    lca.lci()
    lca.lcia()
    return lca.score

def getLCAresults(acts, mymethod):
    
    all_activities = []
    results = []
    for a in acts:
        act = Database('BTC_2019').get(a)
        all_activities.append(act['name'])
        results.append(dolcacalc(act,1,mymethod)) # 1 stays for one unit of each process
        #print(act['name'])
     
    results_dict = dict(zip(all_activities, results))
    
    return results_dict
```

```python
results_BTC_2019 = []
for m in mylist:
    results_all_acts = getLCAresults(acts,m) # total impact per tech
    results_BTC_2019.append(results_all_acts)
```

```python
methods_names = []
for m in mylist:
    m_name = ' '.join(m)
    methods_names.append(m_name)
```

```python
my_output = pd.DataFrame(results_BTC_2019, index=methods_names)
```

```python
print(my_output['Bitcoin Network'])
```

```python
my_output.to_csv('BTC-results_2019.csv', sep = ';')
```
